```python
import matplotlib.pyplot as plt
import numpy as np
import math
import tensorflow as tf
```

```python
def PlotActivationFunction(x, y1, y2, title="Activation y1 and y2"):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15,7))
    fig.suptitle(title)

    ax1.plot(x, y1, '-ro')
    ax1.set(xlabel='x', ylabel='y1')
    ax1.grid()
    ax2.plot(x, y2, '-go')
    ax2.set(xlabel='x', ylabel='y2')
    ax2.grid()

    plt.show()
```

```python
x = np.linspace(-10, 10, 50)
sigmoid = lambda z: 1 / (1 + np.exp(-z))
relu = lambda z: max(0.0, z)
leakyrelu = lambda z, alpha: z if z > 0 else alpha * z
tanh = lambda z: (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
swish = lambda z: z * (1 / (1 + np.exp(-z)))
ELU = lambda z: z if z > 0 else (1 * (np.exp(z) - 1))
```

```python
y_sigmoid_manual, y_sigmoid_lib = sigmoid(x), tf.keras.activations.sigmoid(x)
PlotActivationFunction(x, y_sigmoid_manual, y_sigmoid_lib, "Sigmoid Activation Manually and Library")
```

```python
y_relu_manual, y_relu_lib = [relu(i) for i in x], tf.keras.activations.relu(x)
PlotActivationFunction(x, y_relu_manual, y_relu_lib, "ReLU Activation Manually and Library")
```

```python
y_leakyrelu_manual, y_leakyrelu_lib = [leakyrelu(i, 0.1) for i in x], tf.keras.layers.LeakyReLU(alpha=0.1)(x)
PlotActivationFunction(x, y_leakyrelu_manual, y_leakyrelu_lib, "Leaky ReLU Activation Manually and Library")
```

```python
y_tanh_manual, y_tanh_lib = tanh(x), tf.keras.activations.tanh(x)
PlotActivationFunction(x, y_tanh_manual, y_tanh_lib, "Tanh Activation Manually and Library")
```

```python
y_swish_manual, y_swish_lib = swish(x), tf.keras.activations.swish(x)
PlotActivationFunction(x, y_swish_manual, y_swish_lib, "Swish Activation Manually and Library")
```

```python
y_elu_manual, y_elu_lib = [ELU(i) for i in x], tf.keras.activations.elu(x)
PlotActivationFunction(x, y_elu_manual, y_elu_lib, "ELU Activation Manually and Library")
```

```python
def PlotDerivative(x, y, dydx, title="Functions and Its Derivatives"):
    fig, ax = plt.subplots(figsize=(15, 3))
    ax.plot(x, y, color='red', linewidth=3, label="function")
    ax.plot(x, dydx, color='green', linewidth=3, label="derivative")
    ax.legend(loc="upper right")
    plt.axhline(0, color='y', alpha=0.7)
    plt.axvline(0, color='y', alpha=0.7)
    plt.grid()
    plt.title(title)
    plt.show()

x = np.arange(-5.0, 5.0, 0.01)
dx = x[1] - x[0]
```

```python
y_sigmoid = tf.keras.activations.sigmoid(x)
dydx_sigmoid = np.gradient(y_sigmoid, dx)
PlotDerivative(x, y_sigmoid, dydx_sigmoid, title="Sigmoid Function and Its Derivative")

```

```python
y_relu = tf.keras.activations.relu(x)
dydx_relu = np.gradient(y_relu, dx)
PlotDerivative(x, y_relu, dydx_relu, title="ReLU Function and Its Derivative")

```

```python
y_leakyrelu = tf.keras.layers.LeakyReLU(alpha=0.3)(x)
dydx_leakyrelu = np.gradient(y_leakyrelu, dx)
PlotDerivative(x, y_leakyrelu, dydx_leakyrelu, title="Leaky ReLU Function and Its Derivative")

```

```python
y_tanh = tf.keras.activations.tanh(x)
dydx_tanh = np.gradient(y_tanh, dx)
PlotDerivative(x, y_tanh, dydx_tanh, title="Tanh Function and Its Derivative")

```

```python
y_swish = tf.keras.activations.swish(x)
dydx_swish = np.gradient(y_swish, dx)
PlotDerivative(x, y_swish, dydx_swish, title="Swish Function and Its Derivative")

```

```python
y_elu = tf.keras.activations.elu(x)
dydx_elu = np.gradient(y_elu, dx)
PlotDerivative(x, y_elu, dydx_elu, title="ELU Function and Its Derivative")
```
