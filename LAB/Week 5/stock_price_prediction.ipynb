```python
ma100 = df.Close.rolling(100).mean()
ma100
```

```python
ma200 = df.Close.rolling(200).mean()
ma200
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.Close, label = f'{stock} Close Price', linewidth = 2)
plt.plot(ma100, label = f'{stock} M100', linewidth = 1)
plt.plot(ma200, label = f'{stock} M200', linewidth = 1)
plt.title(f'{stock} Volume over time')
plt.legend()
plt.show()
```

```python
ema100 = df.Close.ewm(span=100, adjust = False).mean()
```

```python
ema200 = df['Close'].ewm(span=200, adjust = False).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df.Close, label = f'{stock} Close Price', linewidth = 2)
plt.plot(ema100, label = f'{stock} Exp.M100', linewidth = 1)
plt.plot(ema200, label = f'{stock} Exp.M200', linewidth = 1)
plt.title(f'{stock} Volume over time')
plt.legend()
plt.show()
```
