# ML **Project**

# Load **data**


```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

## Data preparation


### Split to X and Y


```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1) #axis =1 cot =0 hang
x
```

### Split data to train

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_train #đưa dữ liệu dầu vào để train
```

```python
x_test #dữ liệu dùng để test
```

## Model Building

### model building with **linear regression** (PP Hồi quy tuyến tính)

####Train model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

####apply model to a prediction

```python
y_lr_train_pred = lr.predict(x_train) #dự đoán đầu ra dựa trên tập dữ liệu train
y_lr_test_pred = lr.predict(x_test)#dự đoán đầu ra dựa trên tập dữ liệu keiemr tra
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

####Evalute model performance(đánh giá hiệu suất dựa trên mô hình LR)

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred) # so sánh giá triij ytrain với giá trị dự đoán ylrtrain
lr_train_r2 = r2_score(y_train,y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

### Random forest(Mô hình rừng cây)

####Train

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)## độ sâu của mỗi cây quyết định = 2
rf.fit(x_train, y_train)
```

####Apply

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

#### Evalute

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **Model** **comparison** (So sánh mô hình)

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index()
df_models.drop(['index'], axis=1, inplace=True)
df_models
```

## **Data visualization** (Hình ảnh hóa dữ liệu)

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,5))
plt.scatter(x=y_train, y=y_lr_train_pred, alpha=0.3)
z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train), '#F8766D')
plt.ylabel('Logistic Regression')
plt.xlabel('Experimental logS')
plt.show()
```
